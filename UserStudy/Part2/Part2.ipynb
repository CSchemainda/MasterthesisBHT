```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import matplotlib as mpl

fig = plt.figure(constrained_layout=True, figsize=(20, 6))
spec = mpl.gridspec.GridSpec(ncols=2, nrows=1)
spec.update(wspace=0.2, hspace=0.35)
plt.rc('ytick', labelsize=22)

ax1 = fig.add_subplot(spec[0,0])
ax2 = fig.add_subplot(spec[0,1])


objects = ('very unsure', 'unsure', 'undecided', 'sure', 'very sure')
y_pos = np.arange(len(objects))
performance1 = [0,0,0,4,3]
performance2 = [1,1,2,5,2]

ax1.bar(y_pos, performance1, align='center', alpha=1.0, color='royalblue')
ax1.set_xticks(y_pos, objects, fontsize=18)
ax1.set_ylabel('Number of Participants',labelpad=12, fontsize=20)
ax1.set_xlabel('How sure are you about your assessment?', labelpad=12, fontsize=20)
ax1.set_title('Group 1: No Decomposition Charts', fontsize=26)
ax1.yaxis.set_ticks([0,1,2,3,4,5]) 


ax2.bar(y_pos, performance2, align='center', alpha=1.0, color='royalblue')
ax2.set_xticks(y_pos, objects, fontsize=18)
ax2.set_ylabel('Number of Participants',labelpad=12, fontsize=20)
ax2.set_xlabel('How sure are you about your assessment?',labelpad=12, fontsize=20)
ax2.set_title('Group 2: With Decomposition Charts', fontsize=26)

plt.savefig("StudyTask3.pdf",bbox_inches='tight') 

plt.show()
```
